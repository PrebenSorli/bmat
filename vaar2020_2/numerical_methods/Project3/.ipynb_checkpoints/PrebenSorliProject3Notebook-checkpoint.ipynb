# Problem 2

We want to verify that our Singular Value Decomposition of a matrix A is correct and that $U$, $V$ and $\Sigma$ are indeed matrices satisfying the properties of the SVD. I have chosen to do this in a very basic way, by simply calculating the matrix products $UU^T$, $VV^T$ and $U\Sigma V^T$ and checking that it prints $I$, $I$, $A$ respectively.

```python
import numpy as np
import numpy.linalg as LA

def svdVerifier(A, U, V, Sigma):
    print( np.matmul(U, np.matmul(Sigma, V)) )
    print( np.matmul(U, np.transpose(U)) )
    print( np.matmul(V, np.transpose(V)))
    print( Sigma )

A = np.array([[2, 0 , -1], [0, 1, 0], [-1, 0, 2]])

U = np.array([[1/2, 1/(np.sqrt(2)), 1/2], [1/(np.sqrt(2)), 0, -1/(np.sqrt(2))], [1/2, -1/(np.sqrt(2)), 1/2]])

Sigma = np.array([[1, 0, 0], [0, 3, 0], [0, 0, 1]])

print(A)
print(U)
print(Sigma)

svdVerifier(A, U, U, Sigma)
```

We see that the products turns out as expected.

# Problem 10

Here we implement Euler's method and Heun's method for numerical approximations of ODEs. 

```python
def euler(h, g, y1_0, y2_0, t_0, t_end):
    """ Euler's method for numerical approximation of ODE.
    
    Parameters:
            h: step-size
            g: RHS of our ODE as a function
            y1_0 : Initial value y1(t_0)
            y2_0 : Initial value y2(t_0)
            t_0 : time at start
            t_end : time at final t_n in our discretized grid of times t_n
        Returns:
            yj_list: list of yj-values where y[i] is approx of yj(t_n) j = 1,2
            time_grid: discretization of the time interval [t_0, t_end]
            
    """
    # number of discretized points t_n
    N = int((t_end - t_0) / h)
    # Following the notation in the theory, we have N+1 discrete time values linearly spaced
    time_grid = np.linspace(t_0, t_end, N + 1)
    
    # Initialise array to store y-values
    y1_list = np.zeros(N + 1)
    y2_list = np.zeros(N + 1)
    # Assign initial condition to first element
    y1_list[0] = y1_0
    y2_list[0] = y2_0
    
    Dy1, Dy2 = g[0], g[1]
    
    for i in range(0, N):
        y1_list[i + 1] = y1_list[i] + h * Dy1(y1_list[i], y2_list[i])
        y2_list[i + 1] = y2_list[i] + h * Dy2(y1_list[i], y2_list[i])
    return time_grid, y1_list, y2_list


def heun(h, g, y1_0, y2_0, t_0, t_end):
    """Heun's method for numerical approximation of ODE.
    Also widely known as improved Euler method or the second-order Runge-Kutta method
    
     Parameters:
            h: step-size
            g: RHS of our ODE as a function
            y1_0 : Initial value y1(t_0)
            y2_0 : Initial value y2(t_0)
            t_0 : time at start
            t_end : time at final t_n in our discretized grid of times t_n
        Returns:
            y_list: list of y-values where y[i] is approx of y(t_n)
            time_grid: discretization of the time interval [t_0, t_end]
            
    """
       # number of discretized points t_n
    N = int((t_end - t_0) / h)
    time_grid = np.linspace(t_0, t_end, N + 1) 
    y1_list = np.zeros(N + 1)
    y2_list = np.zeros(N + 1)
    y1_list[0] = y1_0
    y2_list[0] = y2_0
    
    Dy1, Dy2 = g[0], g[1]
    
    for i in range(0, N):
        temp_1 = Dy1(y1_list[i], y2_list[i])
        temp_2 = Dy2(y1_list[i], y2_list[i])
        y1_next = y1_list[i] + 1/2 * h * (temp_1 + Dy1(y1_list[i]+h*temp_1, y2_list[i] + h * temp_2))
        y1_list[i + 1] = y1_next
        
        y2_next = y2_list[i] + 1/2 * h * (temp_2 + Dy2(y1_list[i]+h*temp_1, y2_list[i] + h * temp_2))
        y2_list[i + 1] = y2_next
        
    return time_grid, y1_list, y2_list
    
```

# Problem 11 (d)

In this problem we are solving a Boundary Value Problem numerically. We create the matrices $A_h$ and $G_h$ as in problem $11$, here as Python-functions depending on $M$ the number of discretization points. The differential equation we are solving in problem $11$
    \begin{equation}
        y''(\theta)+\omega^2y(\theta)=0
    \end{equation}
is not too difficult to solve analytically. With boundary values $y(0)=0, y(1)=1$, we get $y=\sin\left(\frac{\pi}{2}t\right)$, which we use as our reference solution.

```python
def A_h(M):
    one = np.array([1 for i in range(M-1)])
    main_diag = np.array([-2 for i in range(M)])
    return np.diag(main_diag) + np.diag(one, -1) + np.diag(one, 1)

def BvPprob(omega, eq, a, b, alpha, beta, M):
    h = (b-a)/(M+1)
    disc_times = np.linspace(a,b, M+1)
    G_h = (1/(h**2))*A_h(M) + omega**2*np.identity(M)
    b = np.array([float(eq[i-1](disc_times[i])) for i in range(1, M+1)])
    k = np.zeros(M)
    k[0] = -alpha/(h**2)
    k[-1] = -beta/(h**2)
    b += k
    return np.linalg.solve(G_h, b)
    
def reference(x):
    return np.sin(np.pi*x/2)

def err(h):
    M = round(1/h) -1
    disc_times = [i*h for i in range(1,M+1)]
    approximation = BvPprob(np.pi/2, [lambda x: 0 for i in range(M)], 0, 1, 0, 1, M)
    reference_vec = list(map(reference, disc_times))
    return np.sqrt(1/M)*np.linalg.norm(reference_vec-approximation)


x2 = lambda x: x**2
plotlist = [1/(i*100) for i in range(1,50)]
funclist = list(map(err, plotlist))
plt.figure()
plt.axes(xlabel = "h", ylabel = "error")
plt.title("Error of the numerical approximation for BVP on [0,1]")
plt.loglog(plotlist, funclist, label = "square error")
plt.loglog(plotlist, x2(np.array(plotlist)), label = "y = x^2")
plt.legend()
plt.show()
```

Here we see that the 2-norm of the error function goes to zero as $\mathcal{O}(h^2)$.
